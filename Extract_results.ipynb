```python
import os
import re
import json
import pandas as pd
import requests
import sys
import glob
import openai
from google import genai  # para la API de Gemini
from google.genai import types
from openai import OpenAI


def call_gpt(prompt, model="gpt-4o-mini-2024-07-18", temperature=0):
    client = OpenAI()
    
    response = client.chat.completions.create(
        model=model,
        messages=[
            {
                "role": "user",
                "content": [
                    {"type": "text", "text": prompt},
                ]
            }
        ],
        max_tokens=8000,
        temperature=temperature,
    )    
    score = response.choices[0].message.content
    
    return score
        
    
```

```python

# txt_folder = "summaries"
# input_dir = 'results_TRN'
output_dir = 'Results_internal/Evaluated'
for file_name in os.listdir(output_dir):
    if file_name.endswith('.csv'):
        output_path = os.path.join(output_dir, file_name)
        
        print(output_path)
        df = pd.read_csv(output_path)
        temperature = 0
        
        for idx, row in df.iterrows():

            analyse = row['value_gemini']
            prompt = f"""Extract only the final score given in the following evaluation.
            Return only the numerical score and no other information. 
            Here is the evaluation from which the score should be extracted:{analyse}"""
            score = call_gpt(prompt)
            df.at[idx, "score_gemini"] = score
            
            analyse = row['value_gpt']
            prompt = f"""Extract only the final score given in the following evaluation.
            Return only the numerical score and no other information. 
            Here is the evaluation from which the score should be extracted:{analyse}"""
            score = call_gpt(prompt)
            df.at[idx, "score_gpt"] = score

            df.to_csv(output_path, index=False)
    
```
